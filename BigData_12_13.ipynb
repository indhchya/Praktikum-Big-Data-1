# Exercise 1: Identifying and Handling Missing Data

```python
import pandas as pd

# Sample dataset with missing values
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', None],
    'Age': [24, 30, None, 22, 35],
    'Salary': [48000, None, 57000, None, 60000]
}
df = pd.DataFrame(data)

# Filling missing values and dropping rows
df['Age'].fillna(df['Age'].mean(), inplace=True)
df['Salary'].fillna(df['Salary'].median(), inplace=True)
df.dropna(subset=['Name'], inplace=True)
print('After cleaning:\n', df)
```

# Exercise 2: Standardizing Categorical Data

```python
# Sample dataset with inconsistent categorical values
data = {
    'Product': ['Laptop', 'Laptop', 'Desktop', 'Tablet', 'Tablet'],
    'Category': ['Electronics', 'electronics', 'Electronics', 'Gadgets', 'gadgets']
}
df = pd.DataFrame(data)

# Standardize category values
df['Category'] = df['Category'].str.capitalize()
print('Standardized Data:\n', df)
```

```python
import pandas as pd

df = pd.read_csv('disney_movies.csv')

print("Tipe data kolom 'genre':", df['genre'].dtype)

if 'genre' in df.columns:
    df['genre'] = df['genre'].astype(str)
    df['genre'] = df['genre'].replace(['nan', 'NaN', ''], 'Unknown')
    df ['genre'] = df['genre'].str.lower().str.strip()
    print("Nilai unik setelah standarisasi:", df['genre'].unique())

df = df.drop_duplicates()
print("20 data teratas setelah menghapus duplikat:")
print(df.reset_index(drop=True).head(20))
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from scipy import stats

# Membaca dataset Netflix Movies and TV Shows
df = pd.read_csv('Netflix_Movies_and_TV_Shows.csv')

# Menampilkan informasi awal tentang dataset
print(df.info())
print(df.describe())
print(df.isnull().sum())

# Mengisi nilai null dengan rata-rata (hanya untuk kolom numerik)
df.fillna(df.mean(numeric_only=True), inplace=True)

numerical_columns = ['release_year', 'rating_score', 'duration_minutes']
numerical_columns = [col for col in numerical_columns if col in df.columns]

# If numerical_columns is empty, handle the situation (e.g., print a message or raise an error)
if not numerical_columns:
    print("Error: No numerical columns found for scaling.")

else:
    scaler = MinMaxScaler()
    df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

    z_scores = stats.zscore(df[numerical_columns])
    df = df[(z_scores < 3).all(axis=1)]

    # Menampilkan hasil akhir setelah preprocessing
    print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset Netflix Movies and TV Shows
# Pastikan file CSV tersedia di direktori kerja Anda
df = pd.read_csv('Netflix_Movies_and_TV_Shows.csv')

plt.figure(figsize=(12, 10))

# Subplot 1: Distribusi Rating
plt.subplot(2, 2, 1)
if 'rating' in df.columns:
    sns.histplot(df['rating'].dropna(), bins=30, kde=False)  # Drop NaN untuk menghindari error
    plt.title('Distribusi Rating')
    plt.xlabel('Rating')
    plt.ylabel('Frekuensi')
else:
    plt.text(0.5, 0.5, "Kolom 'rating' tidak ditemukan", ha='center', va='center')

# Subplot 2: Release Year vs Type (Movie/TV Show)
plt.subplot(2, 2, 2)
if 'release_year' in df.columns and 'type' in df.columns:
    sns.boxplot(x='type', y='release_year', data=df.dropna(subset=['release_year', 'type']))
    plt.title('Release Year vs Type')
    plt.xlabel('Type (Movie/TV Show)')
    plt.ylabel('Release Year')
else:
    plt.text(0.5, 0.5, "Kolom 'release_year' atau 'type'", ha='center', va='center')

# Subplot 3: Scatterplot Durasi vs Release Year
plt.subplot(2, 2, 3)
if 'duration_minutes' in df.columns and 'release_year' in df.columns:
    sns.scatterplot(x='duration_minutes', y='release_year', data=df.dropna(subset=['duration_minutes', 'release_year']))
    plt.title('Durasi vs Release Year')
    plt.xlabel('Durasi (Menit)')
    plt.ylabel('Release Year')
else:
    plt.text(0.5, 0.5, "Kolom 'duration_minutes' atau 'release_year'", ha='center', va='center')

# Subplot 4: Pie Chart Genre
plt.subplot(2, 2, 4)
if 'genre' in df.columns:
    genre_counts = df['genre'].value_counts().head(5)  # Tampilkan top 5 genre
    plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%')
    plt.title('Diagram Lingkaran Genre (Top 5)')
else:
    plt.text(0.5, 0.5, "Kolom 'genre'", ha='center', va='center')

plt.tight_layout()
plt.show()
```
